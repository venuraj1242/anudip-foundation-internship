## Lab1: Visualize the daily temperature changes over time in a city and give your conclusion

```python
import matplotlib.pyplot as plt
days = list(range(1, 32))
temperature = [65, 68, 70, 72, 75, 76, 78, 80, 81, 79, 75, 72, 70, 68, 67, 69, 70, 73, 75, 76, 78, 80, 81, 82, 83, 82, 80, 78, 76, 74, 71]
plt.plot(days, temperature, color='blue', marker='o', linestyle='-')
plt.xlabel('Day of the Month')
plt.ylabel('Temperature (°F)')
plt.title('Daily Temperature Changes in July')
plt.grid(True)
plt.xlim(1, 31)  
plt.ylim(60, 90)  
plt.show()
```

Conclusion:
The temperature in July generally increases from the beginning of the month to the middle of the month, then decreases towards the end of the month. The highest temperature is 83°F, and the lowest temperature is 65°F.
