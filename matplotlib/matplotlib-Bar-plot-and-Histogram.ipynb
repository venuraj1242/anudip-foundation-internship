## Basic Plotting

```python
import matplotlib.pyplot as plt
# Data
x = [1, 2, 3, 4, 5,6,7]
y = [10, 20, 25, 30, 35,21,37]
# Create a line plot
plt.plot(x, y)
# Add labels and title
plt.xlabel("month")
plt.ylabel("salary in k")
plt.title("salay")
# Show the plot
plt.show()
```

## line plotting

Continuous Data: Line plots are particularly useful for visualizing continuous or
sequential data, such as time series data, where data points are ordered chronologically
or by some other continuous variable.
Line plots are useful for various applications, including:
●
Visualizing stock price changes over time.
●
Displaying temperature variations during a day or year.
●
Analyzing trends in website traffic.
●
Monitoring changes in sensor data.
●
Comparing data series, such as sales figures for different products


```python
import matplotlib.pyplot as plt
days = list(range(1, 31))
temperatures = [68, 70, 72, 75, 77, 80, 82, 83, 81, 78, 75, 72, 71, 70, 72, 74, 77, 79,
80, 82, 84, 86, 88, 87, 85, 82, 80, 77, 75, 73]
# Create the line plot
plt.plot(days, temperatures,marker='o')
# Add labels and a title
plt.xlabel("Day of the Month")
plt.ylabel("Temperature (°F)")
plt.title("Daily Temperature Over a Month")
plt.show()
```

 ### Bar plots are commonly used to illustrate various types of data, such as:
●
Comparing sales figures for different products or regions.
●
Showing the distribution of age groups in a population.
●
Visualizing the frequency of specific events or outcomes.
●
Representing survey responses by category.

```python
genres = ["Mystery", "Romance", "Science Fiction", "Fantasy", "Thriller"]
books_sold = [120, 90, 80, 110, 70]
# Create the bar plot
plt.bar(genres, books_sold)
# Add labels and a title
plt.xlabel("Genre")
plt.ylabel("Number of Books Sold")
plt.title("Books Sold by Genre Over a Year")
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Sample data: Population distribution by age group in a city
age_groups = ["0-10", "11-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71+"]
population = [15000, 22000, 30000, 28000, 25000, 18000, 12000, 8000]
# Create the bar plot
plt.barh(age_groups, population, color='lightblue')
# Add labels and a title
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.title("Population Distribution by Age Group")
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Line Plot

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Curve')
plt.grid(True)
plt.show()
```

2. Scatter Plot
A scatter plot is used to visualize the relationship between two numerical variables. Each data point is represented by a dot.

```python
# Sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Create the plot
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

3. Bar Plot
A bar plot is used to represent categorical data with rectangular bars.

```python
# Sample data
categories = ['A', 'B', 'C', 'D']
values = [25, 35, 20, 30]

# Create the plot
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

4. Histogram
A histogram is used to represent the distribution of numerical data.

```python
# Sample data
data = np.random.randn(1000)

# Create the plot
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

5. Pie Chart
A pie chart is used to represent categorical data as slices of a pie.

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 30]

# Create the plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio for a circular pie chart
plt.title('Pie Chart')
plt.show()
```
