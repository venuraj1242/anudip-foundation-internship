 ## Lab1: 
 Analyze the relationship between the size of houses (measured in square
 footage) and their selling prices in a particular neighborhood. You have collected
 data on various houses in that neighborhood.Create a scatter plot using the
 below data and share your conclusion/analysis.
 Input:
 square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800,
 3000])
 selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

```python
import matplotlib.pyplot as plt
square_footage = [1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000]
selling_prices = [250, 290, 315, 380, 410, 450, 500, 525, 570, 610]
plt.scatter(square_footage, selling_prices)
plt.xlabel("Square Footage")
plt.ylabel("Selling Price")
plt.title("House Size vs. Selling Price")
plt.show()
```

##  Lab4: 
 Suppose you're a sales manager for an e-commerce company, and you
 want to create a figure with subplots to compare the sales performance of
 different product categories over time. You have sales data for four product
categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share
 your conclusion/analysis.
 Input:
 months = np.arange(1, 13)
 electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000,
 36000, 38000, 39000, 41000, 42000])
 clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
 23000, 24000, 25000, 26000])
 home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
 25000, 26000, 27000, 28000, 29000])
 sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
 19000, 20000, 21000, 22000, 23000])

```python
import matplotlib.pyplot as plt
import numpy as np
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Sales Performance by Product Categories')
axs[0, 0].plot(months, electronics_sales, marker='o', color='blue', linestyle='-')
axs[0, 0].set_title('Electronics')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sales Amount')
axs[0, 1].plot(months, clothing_sales, marker='o', color='green', linestyle='-')
axs[0, 1].set_title('Clothing')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Sales Amount')
axs[1, 0].plot(months, home_garden_sales, marker='o', color='red', linestyle='-')
axs[1, 0].set_title('Home & Garden')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Sales Amount')
axs[1, 1].plot(months, sports_outdoors_sales, marker='o', color='purple', linestyle='-')
axs[1, 1].set_title('Sports & Outdoors')
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Sales Amount')

plt.tight_layout()

plt.show()
```

## ChatGPT exercise
 Suppose you work for a retail company, and you have dummy data containing sales
 data for the past year. The data includes information such as SalesDate,product
names,regions, sales quantities, prices, and dates. You have to generate a bar chart
 ,pie plot on region and prices and line chart on SalesDate and prices columns.
 Further, you need to get some inference out of the chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {'SalesDate': pd.date_range(start='2023-01-01', periods=365),
        'Product': np.random.choice(['Product A', 'Product B', 'Product C'], 365),
        'Region': np.random.choice(['North', 'South', 'East', 'West'], 365),
        'Quantity': np.random.randint(1, 50, 365),
        'Price': np.random.uniform(10, 100, 365)}
df = pd.DataFrame(data)
region_sales = df.groupby('Region')['Price'].sum()
region_sales.plot(kind='bar')
plt.title('Total Sales by Region')
plt.ylabel('Total Sales')
plt.show()
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.show()
sales_over_time = df.groupby('SalesDate')['Price'].sum()
sales_over_time.plot(kind='line')
plt.title('Sales Over Time')
plt.ylabel('Total Sales')
plt.xlabel('Date')
plt.show()
```

#### Analysis:
Bar chart: This chart shows the total sales for each region. We can identify which region contributes the most to overall sales.

Pie chart: This chart shows the percentage contribution of each region to total sales. We can easily compare the market share of different regions.

Line chart: This chart shows the sales trend over time. We can identify peak sales periods, seasonal patterns, and overall sales growth.

#### Possible insights:

The region with the highest sales is (region name) and contributes (percentage)% to total sales.
There is a (increasing/decreasing) trend in overall sales over the past year.
Sales are typically higher in (specific months or seasons).
Some regions might show more significant sales fluctuations than others.
